Getting Data For the whole Page Using For Loops

```python
#create empty lists for data points 
name = []
mileage = []
dealer_name = []
rating_count = []
rating = []
price = []


for result in results:
  #name
  try:
    name.append(result.find('h2').get_text())
  except:
    name.append("null")
  #mileage
  try:
    mileage.append(result.find('div', {'class':"mileage"}).get_text())
  except:
    mileage.append("null")
    #dealer name
  try:
    dealer_name.append(result.find('div', {'class': 'dealer-name'}).get_text().strip())
  except:
    dealer_name.append("null")
    #rating_count
  try:
    rating_count.append(result.find('span', {'class': 'sds-rating__link sds-button-link'}).get_text())
  except:
    rating_count.append("null")
    #Price
  try:
    price.append(result.find('span', {'class': 'primary-price'}).get_text())
  except:
    price.append("null")
    #rating
  try:
    rating.append(result.find('span', {'class': 'sds-rating__count'}).get_text())
  except:
    rating.append("null")
```

# **Extracting From Multiple Pages (Pagination)**

```python
#create empty lists for data points 
name = []
mileage = []
dealer_name = []
rating_count = []
rating = []
price = []

   #loop throug the first to 10th web page
for i in range(1,11):
  website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=honda&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip='
  response  = requests.get(website)
  soup = BeautifulSoup(response.content, 'html.parser')
  results = soup.find_all('div', {"class": "vehicle-card"})

  for result in results:
    #name
    try:
      name.append(result.find('h2').get_text())
    except:
      name.append("null")
    #mileage
    try:
      mileage.append(result.find('div', {'class':"mileage"}).get_text())
    except:
      mileage.append("null")
      #dealer name
    try:
      dealer_name.append(result.find('div', {'class': 'dealer-name'}).get_text().strip())
    except:
      dealer_name.append("null")
      #rating_count
    try:
      rating_count.append(result.find('span', {'class': 'sds-rating__link sds-button-link'}).get_text())
    except:
      rating_count.append("null")
      #Price
    try:
      price.append(result.find('span', {'class': 'primary-price'}).get_text())
    except:
      price.append("null")
      #rating
    try:
      rating.append(result.find('span', {'class': 'sds-rating__count'}).get_text())
    except:
      rating.append("null")





   


```

```python
df = pd.DataFrame({'Name': name, 'Price': price, 'Mileage': mileage, 'Dealer_name': dealer_name, 'Rating_count': rating_count, 'Rating': rating})
```

```python
df
```

Data Cleaning

```python
df['Rating_count'] = df['Rating_count'].apply(lambda x:x.strip('reviews)').strip('('))
df['Price'] = df['Price'].apply(lambda x:x.strip('$'))
df['Mileage'] = df['Mileage'].apply(lambda x:x.strip('mi.'))
df
```
